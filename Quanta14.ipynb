```python
import yfinance as yf
import pandas as pd

ticker = "GOLD"  # เปลี่ยนเป็นสัญลักษณ์หุ้นที่ต้องการ
period = "10y"  # เปลี่ยนเป็นระยะเวลาที่ต้องการ
# ดึงข้อมูลย้อนหลัง
data = yf.download(ticker, 
                   start="2014-05-01",  
                   end="2024-05-01", 
                   interval='1mo',)

#มูลค่าทองคำโลก 1 ออนซ์ = 1850.23 USD
data = data*1850.23 

# แปลงข้อมูลเป็น DataFrame
df = pd.DataFrame(data)

# แสดงผล DataFrame บนหน้าจอ
print(df.round(2))
```

```python
dfpl = df[0:120]
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=dfpl.index,
                                     open=dfpl['Open'],
                                     high=dfpl['High'],
                                     low=dfpl['Low'],
                                     close=dfpl['Close']),])
fig.show()
```
